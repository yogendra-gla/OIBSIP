```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data = pd.read_csv('/content/Unemployment in India.csv')
```

```python
data2 = pd.read_csv('/content/Unemployment_Rate_upto_11_2020.csv')
```

```python
data2.head
```

```python
print(data2.isnull().sum())
```

```python
data2.columns = ["States","Dates","Frequency","Estimated Unemployment Rate", "Estimates Employed","Estimated Labour Participation Rate",
                 "Region","longitude","latitude"]
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(data2.corr())
plt.style.use('seaborn-darkgrid')
plt.show
```

```python
plt.title("unemployment in India")
sns.histplot(x="Estimated Unemployment Rate", hue = "Region", data=data2)
plt.show
```

```python
data3=data2[["States","Region","Estimated Unemployment Rate"]]
figure = px.sunburst (data3 , path={"Region","States"},values = "Estimated Unemployment Rate", width = 500, height = 500,title = "unemployment rate in india")
figure.show()
```
